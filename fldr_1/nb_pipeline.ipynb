```python
%sql
create catalog if not exists dlt_catalog
```

```python
%sql
create schema if not exists dlt_catalog.dlt_schema
```

```python
%sql
create table if not exists dlt_catalog.dlt_schema.dlt_table
(
    id int,
    name string,
    salary double
)
```

```python
%sql
insert into dlt_catalog.dlt_schema.dlt_table
values
(1, 'John', 100000),
(2, 'Jane', 120000),
(3, 'Mary', 150000)
```

```python
%sql
select * from dlt_catalog.dlt_schema.dlt_table
```

```python
%sql
create table if not exists dlt_catalog.dlt_schema.dlt_table_m
(
    id int,
    name string,
    salary double
)
```

```python
%sql
insert into dlt_catalog.dlt_schema.dlt_table_m
values
(1, 'John', 100000),
(2, 'Jane', 120000),
(3, 'Mary', 150000)
```

```python
%sql
create table dlt_catalog.dlt_schema.customer
(
    id int,
    name string,
    order_date date
)
```

```python
%sql
insert into dlt_catalog.dlt_schema.customer
values
(1, 'John', '2022-01-01'),
(2, 'Jane', '2022-01-02'),
(3, 'Mary', '2022-01-03');
```

```python
%sql
insert into dlt_catalog.dlt_schema.customer
values
(4, 'John', '2022-01-04'),
(5, 'Jane', '2022-01-05'),
(6, 'Mary', '2022-01-06');
```

```python
%sql
describe extended dlt_catalog.dlt_schema.customer
```

```python
lld = spark.sql("select max(order_date) from dlt_catalog.dlt_schema.customer").collect()[0][0]
```

```python
type(lld)
```

```python
df = spark.sql(f"""
               select * from dlt_catalog.dlt_schema.customer
               where order_date < '{lld}'
               """)
```
